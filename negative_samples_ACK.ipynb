```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
df = pd.read_csv('acute_kidney_data.csv', dtype={"value": str})
```

```python
df.head(5)
```

```python
df = df[pd.to_numeric(df['value'], errors='coerce').notnull()] # read
df['value'] = pd.to_numeric(df["value"], downcast="float")
```

```python
df.head(5)
```

```python
print(len(df.label.unique()))
```

```python
df = df.drop(columns = ['flag'])
```

```python
df.head(10)
```

```python
df_ = df.pivot_table(values='value', index=['subject_id', 'anchor_age', 'gender'], columns='label', aggfunc=np.mean, fill_value=0)
```

```python
df_.head(10)
```

```python
df_ = df_.reset_index()
```

```python
df_.head(10)
```

```python
df_.shape
```

```python
for x in df_:
    df_.drop(df_.loc[df_[x]==0.00].index, inplace=True)
```
